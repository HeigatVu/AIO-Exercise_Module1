```python
def validate_value(tp, fp, fn):
    if not isinstance(tp, int):
        return "tp must be int"
    elif not isinstance(fp, int):
        return "fp must be int"
    elif not isinstance(fn, int):
        return "fn must be int"
    elif tp <= 0 or fp <= 0 or fn <= 0:
        return "tp, fp, fn must be greater than zero"
    return True
```

```python
def calculate_f1_score(tp, fp, fn):
    validation = validate_value(tp, fp, fn)
    if validation == True:
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f1_score = 2 * ((precision * recall) / (precision + recall))
        return f"precision = {precision}, recall = {recall}, F1-score = {f1_score}"
    return validation
```

```python
calculate_f1_score(tp = 2 , fp = "a", fn = 4)
```

```python
calculate_f1_score(tp = 2 , fp = 3, fn = "a")
```

```python
import math
```

```python
def is_number(n):
    try:
        float(n)
    except ValueError:
        return False
    return True
```

```python
def validate_activation_function(activation_name):
    if activation_name == "sigmoid" or activation_name == "relu" or activation_name == "elu":
        return True
    return False
```

```python
def calculate_activation_function(x, activation_function, alpha = 0.01):
    
    validate_activation = validate_activation_function(activation_function)
    check_type = is_number(x)
    
    if validate_activation and check_type:
        
        if activation_function == "sigmoid":
            sigmoid = 1 / (1 + math.e**(-x))
            return f"{activation_function}: f({x}) = {sigmoid}"
        elif activation_function == "relu":
            if x <= 0:
                relu = 0
            relu = x
            return f"{activation_function}: f({x}) = {relu}"
        elif activation_function == "elu":
            if x <= 0:
                elu = alpha * (math.e**(x) - 1)
            elu = x
            return f"{activation_function}: f({x}) = {elu}"
        
    elif validate_activation == False:
        return f"{activation_function} is not supported"
    elif check_type == False:
        return f"x must be a number"
```

```python
calculate_activation_function(x= 1.5, activation_function= "sigmoid")
```

```python
calculate_activation_function(x= "abc", activation_function= "sigmoid")
```

```python
calculate_activation_function(x= 1.5, activation_function= "belu")
```

```python
calculate_activation_function(x= 3, activation_function= "sigmoid")
```

```python
import math
import random
```

```python
def compute_loss(loss_name):
    target = random.uniform(0, 10)
    pred = random.uniform(0, 10)
    if loss_name == "MAE":
        mae = math.abs(target - pred)
        return (pred, target, mae)
    elif loss_name == "MSE" or loss_name == "RMSE":
        mse = (target - pred)**2
        return (pred, target, mse)
```

```python
def compute_error(num_sample, loss_name):
    n = num_sample
    total_loss = 0
    error = 0
    for i in range(0, n):
        compute_loss_value = compute_loss(loss_name)
        total_loss += compute_loss_value[2]
        print(f"loss name: {loss_name}, sample: {i}, pred: {compute_loss_value[0]}, target: {compute_loss_value[1]}, loss: {compute_loss_value[2]}")
    
    if loss_name == "MAE":
        error += 1/n * total_loss
        print(f"Error: {error}")
    if loss_name == "MSE":
        error += 1/n * total_loss
        print(f"Error: {error}")
    if loss_name == "RMSE":
        error += math.sqrt(1/n * total_loss)
        print(f"Error: {error}")
```

```python
compute_error(5, loss_name= "RMSE")
```

```python
def is_integer(n):
    if isinstance(n, int) and n > 0:
        return True
    return False
```

```python
def approx_sin(rad, n):
    check_iteration = is_integer(n)
    
    if check_iteration:
        result = rad
        factorial = 1
        for i in range(1, n + 1):
            numerator = ((-1)**i) * (rad**(2*i + 1))
            factorial = factorial * (2*i) * (2*i + 1)
            denominator = factorial
            result += numerator / denominator
        return result
    return "n must be integer"
```

```python
approx_sin(rad= 1, n= 10)
```

```python
def approx_cos(rad, n):
    check_iteration = is_integer(n)
    
    if check_iteration:
        result = 1
        factorial = 1
        for i in range(1, n + 1):
            numerator = ((-1)**i) * (rad**(2*i))
            factorial = factorial * (2*i - 1) * (2*i)
            denominator = factorial
            result += numerator / denominator
        return result
    return "n must be integer"
```

```python
approx_cos(rad= 3.14, n= 10)
```

```python
def approx_sinh(rad, n):
    check_iteration = is_integer(n)
    
    if check_iteration:
        result = rad
        factorial = 1
        for i in range(1, n + 1):
            numerator = (rad**(2*i + 1))
            factorial = factorial * (2*i) * (2*i + 1)
            denominator = factorial
            result += numerator / denominator
        return result
    return "n must be integer"
```

```python
a = round(approx_sinh(rad= 3.14, n= 10), 2)
print(a)
```

```python
def approx_cosh(rad, n):
    check_iteration = is_integer(n)
    
    if check_iteration:
        result = 1
        factorial = 1
        for i in range(1, n + 1):
            numerator = (rad**(2*i))
            factorial = factorial * (2*i - 1) * (2*i)
            denominator = factorial
            result += numerator / denominator
        return result
    return "n must be integer"
```

```python
approx_cosh(rad= 3.14, n= 10)
```

```python
def md_nre_single_sample(y, y_hat, n, p):
    y_nroot = y**(1/n)
    y_hat_nroot = y_hat**(1/n)
    result = (y_nroot - y_hat_nroot)**p
    return result
```

```python
print(md_nre_single_sample(y= 100, y_hat= 99.5, n= 2, p= 1))
print(md_nre_single_sample(y= 50, y_hat= 49.5, n= 2, p= 1))
print(md_nre_single_sample(y= 20, y_hat= 19.5, n= 2, p= 1))
print(md_nre_single_sample(y= 0.6, y_hat= 0.1, n= 2, p= 1))
```
